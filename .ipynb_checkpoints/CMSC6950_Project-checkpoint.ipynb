# Step 1: Combine datasets into one final dataset

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
with open ("co2_mm_gl_space.txt") as file1:
    time=[]
    co2=[]
    for line in file1:
        #y=line[0:4]
        #m=line[8:10]
        #ym=m+'/'+y
        time.append(line[13:22])
        #time.append(ym)
        avg=line[25:32]
        co2.append(avg)
        timeco2=[time,co2]
            
            
```

```python
with open ("Temperature.txt") as file2:
    temp=[]
    for i, line in enumerate(file2):
        if i>=347:
            temp.append(line[12:21])
        if i>=891:
            break
```

```python
with open ("sea_extent.txt") as file3:
    ice=[]
    for i, line in enumerate(file3):
        if i>=2:
            ice.append(line[9:-1])
        if i>=546:
            break
```

```python
with open ("phdata.txt") as file4:
    ph=[]
    timeph=[]
    for i, line in enumerate(file4):
        ph.append(line[:6])
```

```python
len(ph)
```

```python
for i in range(len(ph)):

    # replace hardik with shardul
    if ph[i] == 'NaN\t19':
       ph[i] = ''
    if ph[i] == 'NaN\t20':
       ph[i] = ''

ph1=[]
for string in (ph):
    if string == '':
        string = np.nan
    ph1.append(string)
```

```python
co2 = [x.strip(' ') for x in co2]
times=[]
for i in time:
    j = i.replace(' ','')
    times.append(j)

co2.pop(0)
times.pop(0)
```

```python
times2 = [float(string) for string in times]
co2_2=[float(string) for string in co2]
temp_2 = [float(string) for string in temp]
ice_2 = [float(string) for string in ice]
ph_2 = [float(string) for string in ph1]
```

```python
len(ph_2)
```

```python
n = 12.0
year=times2[::12]
tempyear = [sum(temp_2[i:i+12])/n for i in range(0,len(temp_2),12)]
co2year = [sum(co2_2[i:i+12])/n for i in range(0,len(co2_2),12)]
iceyear = [sum(ice_2[i:i+12])/n for i in range(0,len(ice_2),12)]

year = year[:-1]
tempyear = tempyear[:-1]
co2year = co2year[:-1]
iceyear = iceyear[:-1]
#sum(temp_2[:12])/12
```

```python
#monthly dataset
dict = {'Time': times2, 'CO2': co2_2, 'Temp': temp_2, 'IceArea': ice_2, 'pH': ph_2}

df = pd.DataFrame(dict)
df.to_csv('ClimateData.txt', sep='\t',index=False,header=False)
```

```python
#yearly dataset
dict = {'Year': year, 'CO2': co2year, 'Temp': tempyear, 'IceArea': iceyear}

df2 = pd.DataFrame(dict)
df2.to_csv('ClimateDataAnnual.txt', sep='\t',index=False,header=False)
```

```python
#ph only dataset
dfph1={'Time': times2, 'ph': ph_2} #make dictionary of time and ph
dfph = pd.DataFrame(dfph1) #make dataframe
dfph.dropna(inplace=True) #remove rows with NaN values
#write the ph and times to lists with no NaN values
phs = dfph['ph'].tolist() 
timeph = dfph['Time'].tolist()
```

# Step 2: Initial Plots

```python
plt.plot(times2,co2_2, label="monthly")
plt.plot(year,co2year, label="annually")
plt.xlabel('Year')
plt.ylabel('$CO_{2}$ (ppm)')
#plt.setp(plot.axes.get_xticklabels()[::10], visible=True)
plt.xlim(1978,2025)
plt.ylim(330,430)
plt.title("$CO_{2}$ Emissions Over Time")
plt.legend()
plt.savefig('co2.png', dpi=300, bbox_inches='tight')
```

```python
plt.plot(times2,temp_2, label="monthly")
plt.plot(year,tempyear, label="annually")
plt.xlabel('Year')
plt.ylabel('Temperature ($\degree$C)')
plt.xlim(1978,2025)
plt.ylim(10,20)
plt.title("Global Average Temperature Over Time")
plt.legend()
plt.savefig('temp.png', dpi=300, bbox_inches='tight')
```

```python
plt.plot(times2,ice_2,label="monthly")
plt.plot(year,iceyear,label="annually")
plt.xlabel('Year')
plt.ylabel('Arctic Sea Ice Area ($10^6$km$^2$)')
#plt.setp(plot.axes.get_xticklabels()[::10], visible=True)
plt.xlim(1978,2025)
plt.ylim(0,18)
plt.title("Arctic Sea Ice Area Over Time")
plt.legend()
plt.savefig('sea_ice.png', dpi=300, bbox_inches='tight')
```

```python
plt.plot(timeph,phs,label="interpolated")
plt.plot(times2,ph_2,label="original")
plt.xlabel('Year')
plt.ylabel('Ocean pH')
#plt.setp(plot.axes.get_xticklabels()[::10], visible=True)
#plt.xlim(1978,2025)
#plt.ylim(0,18)
plt.title("Ocean pH Over Time")
plt.legend()
plt.savefig('ph.png', dpi=300, bbox_inches='tight')
```
